Import the relevant libraries

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

Load the Data

```python
#raw_data = pd.read_csv('Admittance.csv')
```

```python
raw_data
```

```python
data = raw_data.copy()
data['Admitted'] = data['Admitted'].map({'Yes':1,'No':0})
```

```python
data
```

Declare Variables

```python
y = data['Admitted']
x1 = data['SAT']
```

Let's plot the Data (Scatter plot)

```python
plt.scatter(x1,y,color='C0')
plt.xlabel('SAT',fontsize=20)
plt.ylabel('Admitted',fontsize=20)
plt.show()
```

From the above, those students above in the graph were admitted why below were not

LETS PLOT WITH A REGRESSION LINE

```python
x = sm.add_constant(x1)
reg_lin = sm.OLS(y,x)
results_lin = reg_lin.fit()
plt.scatter(x1,y,color='C0')
y_hat = x1*results_lin.params[1] + results_lin.params[0]
plt.plot(x1,y_hat,lw=2.5,color='C8')

plt.xlabel('SAT',fontsize=20)
plt.ylabel('Admitted',fontsize=20)
plt.show()
```

```python
#Above violate the rules of non linearity hence we have to use another approach.
```

```python
reg_log = sm.Logit(y,x)
results_log = reg_log.fit()
def f(x,b0,b1):
    return np.array(np.exp(b0*x*b1) / (1+np.exp(b0*x*b1)))
f_sorted = np.sort(f(x1,results_log.params[0],results_log.params[1]))
x_sorted = np.sort(np.array(x1))
plt.scatter(x1,y,color='C0')
plt.xlabel('SAT', fontsize=20)
plt.ylabel('Admitted',fontsize=20)
plt.plot(x_sorted, f_sorted,color='C8')
plt.show()
```

```python
#From the above logistic regression curve,it is bounded by 2 outcomes 1 and 0. whe STA is low the probability of getting addmitted is low while high score has 1 probability.
#aroud 1600 the probability is uncertain.
```
