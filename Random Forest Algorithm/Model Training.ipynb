```python
#Simple Linear Regression . Minimal example
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Generate random input to train on

observations = 10000
xs = np.random.uniform(low=-10, high=10, size=(observations,1))
zs = np.random.uniform(-10,10,(observations,1))
inputs = np.column_stack((xs,zs))
print(inputs.shape)
```

```python
#Create the target we will aim at
```

```python
noise = np.random.uniform(-1,1,(observations,1))
```

```python
targets = 2*xs -3*zs +5 + noise
print(targets.shape)
```

```python
#plot the training data. it show that there is a strong trend for our model to learn and reproduce
```

```python
targets = targets.reshape(observations,1)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(xs, zs,targets)
ax.set_xlabel('xs')
ax.set_ylabel('zs')
ax.set_zlabel('targets')
ax.view_init(azim=100)
plt.show()
targets = targets.reshape(observations,1)
```

```python
#Initialize variables
```

```python
init_range = 0.1
weights = np.random.uniform(-init_range, init_range, size=(2,1))
biases = np.random.uniform(-init_range,init_range,size=1)
print(weights)
print(biases)
```

```python
#from the above 1s and 2nd are weights, 3rd value is biase
```

```python
#Set a Learning rate.
```

```python
learning_rate = 0.02
```

```python
#Train the model
```

```python
for i in range (100):
    output = np.dot(inputs, weights) + biases
    deltas = output - targets
    #deltas is 1000X 1 array
    loss=np.sum(deltas **2)/2/observations
    print(loss)
    #if loss is decreasing our machine learning algorithim is working well
    deltas_scaled=deltas / observations
    weights = weights - learning_rate * np.dot(inputs.T, deltas_scaled)
    biases = biases - learning_rate * np.sum(deltas_scaled)
    
```

```python
#From the above , weights and biases are optimized
```
