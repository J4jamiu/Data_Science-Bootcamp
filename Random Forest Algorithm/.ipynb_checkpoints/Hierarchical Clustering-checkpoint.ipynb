#From the products purchased by the customers in the data wholesale_data segment the data with similar customers together

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('Wholesale+customers+data.csv')
```

```python
#X=dataset.iloc[:, [3,4]].values
```

```python
from sklearn.preprocessing import normalize
data_scaled= normalize(dataset)
data_scaled = pd.DataFrame(data_scaled, columns=dataset.columns)
data_scaled.head()
X=data_scaled
```

```python
#Using the Dendogram to find the optimal number of clusters
```

```python
import scipy.cluster.hierarchy as sch
dendogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show
```

```python
#Training the Hierarchical Clustering model on the dataset
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 2, affinity = 'euclidean', linkage='ward')
y_hc= hc.fit_predict(X)
```

```python
#Visualizing the clusters
```

```python
plt.figure(figsize=(10,7))
```

```python
plt.scatter(X['Milk'], X['Grocery'], c=hc.labels_)
```

#Let’s now see an
example with 500 data points. The following script creates 500 data points with 4
cluster centers.


```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# generating dummy data of 500 records with 4 clusters
features, labels = make_blobs(n_samples=500, centers=4, cluster_std = 2.00)

#plotting the dummy data
plt.scatter(features[:,0], features[:,1] )
```

The following script applies agglomerative hierarchical clustering on the dataset.
The number of predicted clusters is 4.


```python
# performing kmeans clustering using AgglomerativeClustering class
hc_model = AgglomerativeClustering(n_clusters=4, affinity= 'euclidean' , linkage='ward' )
hc_model.fit_predict(features)
```

```python
#The output shows the labels of some of the data points in our dataset. You can see
#that since there are 4 clusters, there are 4 unique labels, i.e., 0, 1, 2, and 3.
```

```python
#To plot the predicted clusters, execute the following script
```

```python
#pring the data points
plt.scatter(features[:,0], features[:,1], c= hc_model.labels_, cmap='rainbow' )
```

```python
#Similarly, to plot the actual clusters in the dataset (for the sake of comparison),
#execute the following script.
```

```python
#print actual datapoints
plt.scatter(features[:,0], features[:,1], c= labels, cmap='rainbow' )
```

see how to cluster the Iris dataset using hierarchical
agglomerative clustering. The following script imports the Iris dataset and displays
the first five rows of the dataset.


```python
import seaborn as sns
iris_df = sns.load_dataset('iris' )
iris_df.head()
```

The following script divides the data into features and labels sets and displays the
first five rows of the labels set

```python
# dividing data into features and labels
features = iris_df.drop(['species' ], axis = 1)
labels = iris_df.filter(['species' ], axis = 1)
features.head()
```

Similarly, the following script applies the agglomerative clustering on the feature
set using the AgglomerativeClustering class from the sklearn.cluster module.


```python
# training Hierarchical clustering model
from sklearn.cluster import AgglomerativeClustering

# training agglomerative clustering model
features = features.values
hc_model = AgglomerativeClustering(n_clusters=3, affinity='euclidean' , linkage='ward' )
hc_model.fit_predict(features)
```

The output above shows the predicted cluster labels for the feature set in the Iris
dataset.

```python
#The predicted clusters are printed via the following script.
```

```python
#pring the data points
plt.scatter(features[:,0], features[:,1], c= hc_model.labels_, cmap='rainbow' )
```

You can also create dendrograms using the feature set using the shc module from
the scipy.cluster.hierarchy library. You have to pass the feature set to the linkage
class of the shc module, and then the object of the linkage class is passed to the
dendrogram class to plot the dendrograms, as shown in the following script

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title('Iris Dendograms' )
dend = shc.dendrogram(shc.linkage(features, method='ward' ))
```
