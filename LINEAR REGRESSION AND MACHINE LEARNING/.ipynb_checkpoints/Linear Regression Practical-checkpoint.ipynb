#MULTICOLLINEARITY:

```python
# It is logical that year and mileage are going to have corelation the older the car the lesser the price, the more the mileage
```

```python
data_cleaned.columns.values
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[['Mileage', 'Year','EngineV']]
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif['features'] = variables.columns
```

```python
vif
```

```python
##When VIF =1 it means no multicollinearity. VIF is between >1<= 5 or 7 no generally accepatbale rule
#From the above lets drop vif 10...
```

```python
data_no_multicollinearity = data_cleaned.drop(['Year'], axis=1)
```

###What most people are wondering is not 'when do we have multicollinearity' as we usually do have some. The better question to ask is 'When can we safely ignore multicollinearity'.


Here's a great article on the topic: 
http://statisticalhorizons.com/multicollinearity

***
Regarding the variance inflation factor method that we employed from StatsModels, you can read the full documentation here: https://www.statsmodels.org/stable/generated/statsmodels.stats.outliers_influence.variance_inflation_factor.html

```python
#Create Dummies Variables with use of GET DUMMIES
```

```python
data_with_dummies = pd.get_dummies(data_no_multicollinearity, drop_first = True)
```

```python
data_with_dummies.head()
```

```python
#Rearrange a bit
```

```python
data_with_dummies.columns.values
```

```python
cols = (['Mileage', 'EngineV', 'log_price', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes'])
```

```python
data_preprocessed = data_with_dummies[cols]
data_preprocessed.head()
```

LINEAR REGRESSION MODEL

```python
#Declare the inputs and targets
```

```python
targets = data_preprocessed['log_price']
inputs = data_preprocessed.drop(['log_price'], axis=1)
```

```python
#Scale the data
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

N.B It is discouraged, but once scaled they loose all dummies variables


```python
#SPLIT THE DATA INTO TRAIN AND TEST
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size=0.2, random_state=365)
```

CREATE REGRESSION

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train, y_hat)
plt.xlabel('Targets (y_train)', size=18)
plt.ylabel('Predictions (y_hat)', size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
#Calculate the R^2
```

```python
reg.score(x_train,y_train)
```

```python
#From the above it shows that the Model is expalining 75% variability of the data
```

```python
#Find Weights and Bias
```

```python
#Bias is contained in reg intercept while weight is contained in reg_coef
reg.intercept_
```

```python
reg.coef_
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns= ['Features'])
reg_summary['Weights']= reg.coef_
reg_summary
```

```python
#Weights Intrpretation for 
#1. Continuous varibales:
#A positive weights shows that as a feature increases in value, so do the log_price and 'Price' respectively
#A negative weight shows that as a feature increases in value log_price and 'Price' decrease

#for Dummy Variables:
#1. A positive weight shows that the respective category (Brand) is more expensive tahn the benchmark(Audi)
#2. A negative weight shows that the respective category(Brand) is less expensive than the benchmark (Audi)
```

```python
#The biger the weight the bigger the impact
```

```python
data_cleaned['Brand'].unique()
```

TESTING

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test, y_hat_test)
plt.xlabel('Targets (y_test)', size=18)
plt.ylabel('Predictions (y_hat_test)', size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
#From the above the variables are scattered at the lower part of teh graph indicating a likely error. we can check the predicted prices to validate
```

```python
df_pf = pd.DataFrame(np.exp(y_hat_test),columns=['Predictions'])
df_pf.head()
```

```python
df_pf['Target'] =np.exp(y_test)
df_pf
```

```python
y_test
```

```python
#There is need to reset the index of the y_test. 
```

```python
y_test = y_test.reset_index(drop=True)
y_test.head()
```

```python
#Override exponential of y_test in df_pf(Target)
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
df_pf['Residual'] = df_pf['Target'] - df_pf['Predictions']
```

```python
df_pf['Differences%'] = np.absolute(df_pf['Residual']/df_pf['Target']*100)
df_pf
```

```python
df_pf.describe()
```
