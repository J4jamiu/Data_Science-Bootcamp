```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
crash_df = sns.load_dataset('car_crashes')
crash_df
```

```python
#DISTRIBUTION PLOTS
```

```python
#sns.distplot
#Type shift and tab keys to see detail information of 'Univaria Distribution'(for one variable)
sns.displot(crash_df['not_distracted'], kde=False, bins=25)
```

```python
#JOINT PLOT:Is used to to compare 2 distribution and plot a scrater plot by deafult
```

```python
#SEABORN PLOT to show those speeding(x axis) and those under alcohol(y axis) influence, also generate a Regression line or (Hexagone or ;kde' Kernel Density Estimate)
sns.jointplot(x='speeding', y='alcohol',data= crash_df, kind ='reg' )
```

```python
#SEABORN PLOT to show those  under alcohol influence, also generate a 'kde' (Kernel Density Estimate)
sns.kdeplot(crash_df['alcohol'] )
```

```python
#PAIR PLOTS: It plots relationships across the entire data frame(df) numerical values
```

```python
sns.pairplot(crash_df)
```

```python
#To make changes, let us load another data frame tips data frame
tips_df = sns.load_dataset('tips')
#To see content of tips_dataframe:
tips_df
```

```python
sns.pairplot(tips_df, hue='sex' ,palette='Purples')
```

```python
#RUG PLOT :Plot marginal distributions by drawing ticks along the x and y axes.

#This function is intended to complement other plots by showing the location
#of individual observations in an unobstrusive way.

sns.rugplot(tips_df['tip'])
```

```python
#STYLING: To make plots look good such as dark, white, darkgrid,whitegrid,ticks)
sns.set_style('whitegrid')

#To change the size
plt.figure(figsize=(8,4))

#Label and lines and other elements; use eithe rof the 3 options, paper(use on jupyter notebook),talk(Presentation) or poster used on Posters)
sns.set_context('paper',font_scale=1.5)

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')

#To remove the y axis horizontal line in the graph
#sns.despine(left=True)

##To remove the x axis vertical line in the graph
#sns.despine(bottom=True)
```

```python
#CATEGORICAL PLOTS: Bar plot, count, box, Violing
```

```python
 sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.median )
```

```python
#COUNT PLOT: Use to count e.g number of males and females
sns.countplot(x='sex', data=tips_df)
```

```python
#BOX PLOTS: It allows comparism of different variables and their quotas
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')
```

```python
#VIOLIN PLOT: 
sns.violinplot(x='day', y='total_bill', data=tips_df,hue='sex', split=True)
```

```python
#STRIP PLOT
plt.figure(figsize=(8,5))
sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=True, hue='sex', dodge=True)
```

```python
#SWARM PLOT: It help to sparate so they are not ovelapped
sns.violinplot(x='day', y='total_bill', data=tips_df)
sns.swarmplot(x='day', y='total_bill', data=tips_df, color='white')
```

```python
#OR 
sns.swarmplot(x='day', y='total_bill', data=tips_df)
```

```python
#PALETTES
plt.figure(figsize=(8,5))
sns.set_style('dark')
sns.set_context('talk')
sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex', palette='magma')
#GO TO MATPLOTLIB website ...color maps to see different colors
#To ensure legend is properly placed(0=top left,1 = upper right, 2= uppe left, 3= lower left, 4=lower right 
plt.legend(loc=0)
```
